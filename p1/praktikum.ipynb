## 10. Pengurutan Jumlah Populasi pada Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries = (
    "Argentina",
    "Bolivia",
    "Brazil",
    "Chile",
    "Colombia",
    "Ecuador",
    "Falkland Islands",
    "French Guiana",
    "Guyana",
    "Paraguay",
    "Peru",
    "Suriname",
    "Uruguay",
    "Venezuela",
)
populations = (
    45076704,
    11626410,
    212162757,
    19109629,
    50819826,
    17579085,
    3481,
    287750,
    785409,
    7107305,
    32880332,
    585169,
    3470475,
    28258770,
)

df = pd.DataFrame(
    {
        "Country": countries,
        "Population": populations,
    }
)
df.sort_values(by="Population", inplace=True)
x_coords = np.arange(len(df))
plt.bar(x_coords, df["Population"], tick_label=df["Country"])
plt.xticks(rotation=90)
plt.ylabel("Population (Millions)")
plt.title("South American Populations")

plt.text(
    0.5,
    0.5,
    "Muhamad Rizal Arfiyan - 22.11.5227",
    fontsize=20,
    color="black",
    ha="center",
    va="center",
    alpha=0.2,
    transform=plt.gcf().transFigure,
    rotation=30,
)

plt.show()
```
