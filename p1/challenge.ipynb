## Challenge 1

```python
import matplotlib.pyplot as plt

beach = (
    "Pantai Drini",
    "Pantai Slili",
    "Pantai Kukup",
    "Pantai Parangtritis",
    "Pantai Krakal",
    "Pantai Indrayanti",
    "Pantai Sepanjang",
)
rates = (20, 10, 15, 10, 20, 15, 10)
colors = ("#87410f", "#fff5dc", "#91c36e", "#e67d0a", "#d22d32", "#0000ff", "#01ff00")
explode = (0.1, 0, 0, 0, 0.1, 0, 0)

plt.title("Best pantai in Yogyakarta")
plt.pie(
    rates,
    labels=beach,
    autopct="%1.1f%%",
    colors=colors,
    explode=explode,
    shadow=True,
)

plt.text(
    0.5,
    0.5,
    "Muhamad Rizal Arfiyan - 22.11.5227",
    fontsize=16,
    color="black",
    ha="center",
    va="center",
    alpha=0.2,
    transform=plt.gcf().transFigure,
    rotation=30,
)

plt.show()
```

## Challenge 2

```python
import matplotlib.pyplot as plt

extracurricular = ("Sepak Bola", "Volly", "Menjahit", "Komputer", "Qiro'a")
student = (60, 50, 30, 150, 10)
colors = ("#87410f", "#fff5dc", "#91c36e", "#e67d0a", "#d22d32")

plt.title("Data Peserta Ekstrakurikuler SMK TUNAS BANGSA")
plt.pie(
    student,
    labels=extracurricular,
    autopct=lambda p: "{:1.1f}% ({:,.0f})".format(p, p * sum(student) / 100),
    colors=colors,
)

plt.text(
    0.5,
    0.5,
    "Muhamad Rizal Arfiyan - 22.11.5227",
    fontsize=16,
    color="black",
    ha="center",
    va="center",
    alpha=0.2,
    transform=plt.gcf().transFigure,
    rotation=30,
)

plt.show()
```

## Challenge 3

```python
import matplotlib.pyplot as plt
import numpy as np

product = ("Buku tulis", "Pensil", "Penggaris", "Pulpen", "Pensil warna", "Tas")
sell = (100, 200, 150, 300, 50, 200)

x_coords = np.arange(len(product))
bars = plt.bar(x_coords, sell, tick_label=product)
plt.xticks(rotation=90)
plt.ylabel("Jumlah Produk")
plt.title("Jumlah Produk Terjual Bulan Juni 2018")

for bar, count in zip(bars, sell):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() / 2,
        str(count),
        ha="center",
        va="center",
        color="white",
        fontsize=10,
    )

plt.text(
    0.5,
    0.5,
    "Muhamad Rizal Arfiyan - 22.11.5227",
    fontsize=16,
    color="black",
    ha="center",
    va="center",
    alpha=0.2,
    transform=plt.gcf().transFigure,
    rotation=30,
)

plt.show()
```
